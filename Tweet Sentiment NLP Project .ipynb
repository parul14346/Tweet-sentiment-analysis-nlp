## Data Cleaning and Pre-processing 

```python
train_df = df_train.copy()
train_df.head(1)
```

```python
test_df = df_test.copy()
```

```python
train_df["tweet"] = train_df['tweet'].astype(str).str.lower()
train_df["tweet"].head(10)
```

## Removing URLs

```python
import re

def remove_urls(text):
    return re.sub(r'http\S+|www\S+', '', text)  
```

```python
train_df["tweet"] = train_df["tweet"].apply(remove_urls)
train_df.head(10) 
```

## Remove Mentions (@user, @brand)

```python
def remove_mentions(text):
    return re.sub(r'@\w+', '', text)
```

```python
train_df['tweet'] = train_df['tweet'].apply(remove_mentions)
# train_df['tweet'].head(20)
```

## Removing Spaces

```python
def normalize_spaces(text):
    return re.sub(r'\s+', ' ', text).strip()

train_df['tweet'] = train_df['tweet'].apply(normalize_spaces)
```

```python
# train_df["tweet"].head(30)
```

## Handling EMojis

```python
!pip install emoji
```

```python
import emoji

def convert_emojis(text):
    return emoji.demojize(text, delimiters=(" ", " "))

train_df['tweet'] = train_df['tweet'].apply(convert_emojis) 
```

## Remove Hashtags Only

```python
def remove_hashtag_symbol(text):
    return re.sub(r'#', '', text)

train_df['tweet'] = train_df['tweet'].apply(remove_hashtag_symbol)
# train_df.head(20)
```

## Replace multiple punctations with single

```python
def normalize_punctuation(text):
    # replace multiple ! or ? with single
    text = re.sub(r'!+', '!', text)
    text = re.sub(r'\?+', '?', text)
    return text

train_df['tweet'] = train_df['tweet'].apply(normalize_punctuation)
```

```python
train_df['tweet'].head(30) 
```

## replace common symbol-only profanity patterns

```python
def normalize_profanity(text):
    return re.sub(r'[\@\#\$\%\&\*]{2,}', ' profanity ', text)

train_df['tweet'] = train_df['tweet'].apply(normalize_profanity) 
# train_df['tweet'].head(30)
```

```python
train_df['tweet'] = train_df['tweet'].apply(normalize_spaces)
```

## Modelling

```python
x = train_df['tweet']
y = train_df['label'] 
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(
    x,
    y,
    test_size=0.2,      
    random_state=42,
    stratify=y           
)
```

```python
print("Train label distribution:")
print(y_train.value_counts(normalize=True))

print("\nValidation label distribution:")
print(y_val.value_counts(normalize=True))
```

## Tf-Idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(
    ngram_range=(1, 2),     
    min_df=3,              
    max_df=0.9,             
    sublinear_tf=True       
)
```

```python
X_train_tfidf = tfidf.fit_transform(x_train)
```

```python
X_val_tfidf = tfidf.transform(x_val)
```

```python
print(X_train_tfidf.shape)
print(X_val_tfidf.shape)
```

## Modelling

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    n_jobs=-1
)

model.fit(X_train_tfidf, y_train)
```

```python
y_val_pred = model.predict(X_val_tfidf)
```

```python
from sklearn.metrics import classification_report, f1_score

print("Weighted F1 Score:", f1_score(y_val, y_val_pred, average='weighted'))
print(classification_report(y_val, y_val_pred))
```

## SVM Model

```python
from sklearn.svm import LinearSVC

svm_model = LinearSVC(
    C=1.0,                
    class_weight='balanced',
    random_state=42)
```

```python
svm_model.fit(X_train_tfidf, y_train)
y_val_pred_svm = svm_model.predict(X_val_tfidf)
```

```python
from sklearn.metrics import classification_report, f1_score

print("Weighted F1 Score (SVM):",
      f1_score(y_val, y_val_pred_svm, average='weighted'))

print(classification_report(y_val, y_val_pred_svm))
```

## tune Logistics Regresssion Model

```python
C_values = [0.05, 0.1, 0.5, 1, 2, 5, 10]
```

```python
results = []

for c in C_values:
    model = LogisticRegression(
        C=c,
        max_iter=1000,
        class_weight='balanced',
        n_jobs=-1
    )
    
    model.fit(X_train_tfidf, y_train)
    y_pred = model.predict(X_val_tfidf)
    
    score = f1_score(y_val, y_pred, average='weighted')
    results.append((c, score))
    
    print(f"C = {c} → Weighted F1 = {score:.4f}")
```

```python
best_c = max(results, key=lambda x: x[1])[0]
print("Best C:", best_c)
```

```python
best_model = LogisticRegression(
    C=best_c,
    max_iter=1000,
    class_weight='balanced',
    n_jobs=-1
)

best_model.fit(X_train_tfidf, y_train)
best_preds = best_model.predict(X_val_tfidf)

from sklearn.metrics import classification_report

print(classification_report(y_val, best_preds))
```

## Threshold tuning

```python
y_val_proba = best_model.predict_proba(X_val_tfidf)[:, 1]
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

thresholds = [0.35, 0.4, 0.45, 0.5]

for t in thresholds:
    y_pred_thresh = (y_val_proba >= t).astype(int)
    
    precision = precision_score(y_val, y_pred_thresh)
    recall = recall_score(y_val, y_pred_thresh)
    f1 = f1_score(y_val, y_pred_thresh, average='weighted')
    
    print(f"Threshold={t:.2f} | Precision={precision:.3f} | Recall={recall:.3f} | Weighted F1={f1:.3f}")
```
